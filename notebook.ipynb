The Nobel Prize has been among the most prestigious international awards since 1901. Each year, awards are bestowed in chemistry, literature, physics, physiology or medicine, economics, and peace. In addition to the honor, prestige, and substantial prize money, the recipient also gets a gold medal with an image of Alfred Nobel (1833 - 1896), who established the prize.

![](Nobel_Prize.png)

The Nobel Foundation has made a dataset available of all prize winners from the outset of the awards from 1901 to 2023. The dataset used in this project is from the Nobel Prize API and is available in the `nobel.csv` file in the `data` folder.

In this project, you'll get a chance to explore and answer several questions related to this prizewinning data. And we encourage you then to explore further questions that you're interested in!

```python
# Loading in required libraries
import pandas as pd
import seaborn as sns
import numpy as np

# Start coding here!
```

```python
# Reading dataset
df = pd.read_csv('data/nobel.csv')
df.head()
```

```python
# Analyze the most common gender and birth country for Nobel Prize winners
top_gender = df['sex'].value_counts().idxmax()
top_country = df['birth_country'].value_counts().idxmax()

top_gender, top_country
```

```python
# Analyze the highest ratio of US-born Nobel Prize winners to total winners in all categories

df['decade'] = (df['year'] // 10) * 10

# Number of Nobel Prizes in all categories won by the United States of America
df_grouped_usa = df[df['birth_country'] == 'United States of America']
df_grouped_usa = df_grouped_usa.groupby(['decade'])['category'].count()

# Number of Nobel Prizes in all categories won in total
df_grouped_total = df.groupby(['decade'])['category'].count()

# The decade with the highest proportion of American-born Nobel Prize winners to total winners in all categories
df_grouped = df_grouped_usa / df_grouped_total
max_decade_usa = df_grouped.idxmax()
max_decade_usa
```

```python
# Decade and Nobel Prize category combination with the highest proportion of female laureates

# Number of Nobel Prizes by Female
df_grouped_female = df[df['sex'] == 'Female']
df_grouped_female = df_grouped_female.groupby(['decade', 'category'])['sex'].count()

# Number of Nobel Prizes by all gender
df_grouped_cat_total = df.groupby(['decade', 'category'])['sex'].count()

# Decade and Nobel Prize category combination with the highest proportion
prop = df_grouped_female / df_grouped_cat_total
prop_max = prop.idxmax()

max_female_dict = { prop_max[0]: prop_max[1] }
max_female_dict
```

```python
# The first woman to receive a Nobel Prize and in category

# Sorting by 'sex' and 'year'
df_woman_price = df.sort_values(['sex', 'year'], ascending=[True, True])[['year', 'category', 'full_name', 'sex']].reset_index(drop=True)

# Select first raw
first_woman_name = df_woman_price.loc[0]['full_name']
first_woman_category = df_woman_price.loc[0]['category']

first_woman_name, first_woman_category
```

```python
# Individuals or organizations who won more than one Nobel Prize throughout the years
repeat_list = []

# Count the number of times they won a prize and sort descending
df_repeat_winner = df.value_counts('full_name').sort_values(ascending=False)

# List to store indexes whose values are greater than 1
repeat_list = [index for index, value in df_repeat_winner.items() if value > 1]
repeat_list
```
